# 数组合并
## 常用的方法

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
```

```python
np.concatenate((a,b)) #连接a,b(默认沿着0轴)
```

```python
np.concatenate((a,b),axis = 1)
```

默认情况下，我们认为：行的方向是0轴（从上到下），列（从左到右）的方向是1轴
当然，对于一维的数组合并时，我们只会把一个数组放到另一个后面（也就是沿着1轴）

## 另一种合并的方法

```python
a1 = np.array([1,2,3,4])
a2 = np.array([5,6,7,8])
```

```python
np.stack((a1,a2)) #增加维度，把a1,a2当作两个元素
```

# 数组元素去重

```python
a = np.array([1,2,1,1,1,4,5,6,7,3,2,2])
np.unique(a) #这个函数也是返回一个新数组，不会改变a
```
