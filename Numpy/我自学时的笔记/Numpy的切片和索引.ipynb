```python
import numpy as np
```

# 对一维数组切片

```python
arr = np.arange(10)
arr
```

```python
arr[5:8] #对数组进行切片 获取5-8位
```

```python
a2 = arr[2:8:2] #获取2-8位，隔2位获取一个数字
a2
```

```python
a2[0] = 100
a2
```

```python
arr
```

## 对切片后的数组进行修改也会影响切片前的数组

```python
a3 = arr[5:8].copy() #把arr中的5-8位复制到了a3中，此时再修改a3不会影响arr
a3[0] = 200
a3
```

```python
arr
```

# 对高位数组进行切片

```python
a4 = np.arange(24).reshape(2,3,4)
a4
```

```python
a4[0] #取出第一个 3x4数组
```

```python
a4[0][0] #第一个3x4数组的第一行
```

```python
a4[0][0][0] #第一个3x4数组的第一行的第一个
```

```python
a4[1,2,3] #同 a4[1][2][3]
```

```python
a4[:,0,1:4] #取出每一个 3x4 数组的第一行的第1-4位
```

在切片时，可以这么简化：
- **获取所有** $:$
- **获取前k位** $:k$
- **获取第k位到最后一位** $k:$

# 布尔索引

```python
a = np.arange(20)
a
```

```python
a > 10 #此时会产生一个bool类型的数组，其中第 i 位表示 a[i] > 10 的运算结果
```

```python
a[a>10] #取出bool数组中全部的true的数 及原数组 a中所有的大于10的数
```

```python
a[(a>5)&(a<14)]
```

```python
def check(a):
    return (a>2)&(a<10)
a[check(a)] #也可以使用函数
```

```python
a = np.arange(24).reshape(2,3,4)
a
```

```python
(a>10)&(a%2 == 1)
```

```python
a[(a>10)&(a%2 == 1)] #对高维数组使用bool索引，得到的是一维数组
```
