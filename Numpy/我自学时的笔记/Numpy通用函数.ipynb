```python
import numpy as np
```

```python
a = np.arange(10)
a
```

```python
np.sqrt(a) #求 a的每一个元素的平方根，比for循环改快
```

```python
np.sin(a) #求 sin函数
```

```python
np.cos(a) #求 cos函数
```

```python
a**2 #每一个数取幂
```

```python
b = np.arange(10,0,-1) #同range()函数
b
```

```python
np.maximum(a,b) #要保证 a,b两个数组符合广播的要求
#此函数会返回一个新的数组其中数组的第 k 位为 a中第 k位和 b中第 k位之间较大的数
```

```python
a - b
```

```python
a1 = a.reshape(2,5)
a1
```

```python
a1.mean(axis = 1)#求一行的均值
```

```python
a1.mean(0) #简写，等同于a1.mean(axis = 0)
```

```python
np.max(a1)
```

```python
np.min(a1,axis = 1)
```

```python
np.std(a1,axis = 1) #求每一行的标准差
```

```python
np.median(a1,axis = 1) #求每一行的中位数
```

# 数学相关函数
- sum 总和
- mean 算数平均值
- median 中位数
- quantile(k) 分位数，其中 $k$ 是 $0-1$ 之间的小数
- max 最大值
- min 最小值
- ptp 极值
- var 方差
- std 标准差

```python
b = np.random.randint(1,10,6)
b
```

```python
c = b > 5
c.all() #判断 bool数组中所有元素是不是都为 True
```

```python
c.any() #判断 bool数组中含不含有 True
```
