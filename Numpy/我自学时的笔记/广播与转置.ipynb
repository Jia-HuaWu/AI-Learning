```python
import numpy as np
```

```python
a = np.arange(10).reshape(2,5)
a
```

```python
a + 10 #此时 a中的所有数字都会 +10   --广播
```

```python
a + a #形状相同的数组运算时，会将对应位置的数运算
```

```python
a * a
```

```python
a1 = np.arange(5)
a1
```

```python
a + a1 #由于 a1的大小恰好是 a的一行所以此时会把 a的每一行与 a1相加
```

```python
a1 = np.arange(4)
a + a1 #此时会报错，因为a a1既不相似也不相互包含
```

```python
a1 = np.arange(2).reshape(2,1)
a1
```

```python
a + a1 #每一列对应相加
```
